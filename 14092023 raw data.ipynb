```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('raw_data.csv')
data
```

```python
## Q1: add information at index 9.
## 210, "Jim Carrey","1/1/1970", 'jim.carry@gmail.com', 'Male', 'Amazon', 'USA', 12349.556, 2365

data.loc[9]=[210, "Jim Carrey","1/1/1970", 'jim.carry@gmail.com', 'Male', 'Amazon', 'USA', 12349.556, 2365]
data.to_csv('raw_data.csv',index=False)
data
```

```python
# Q2: add below details into 3rd and 6th column
#3rd: 'graeme.croisier@gmail.com' 6th:'modesta.batterson@gmail.com'

data['email'].loc[3]='graeme.croisier@gmail.com'
data['email'].loc[6]='modesta.batterson@gmail.com'
data.to_csv('raw_data.csv',index=False)
data
```

```python
# Q3: If any country name is invalid change it to valid country name and fill NAN in country column.

data['Country']=data['Country'].str.replace('UAS','USA')
data
```

```python
data['Country']= data['Country'].fillna(data['Country'].mode()[0])
data
```

```python
# Q4: total income column in same dataframe and that all rows should contain sum of income 1 and 2

#removing dollar sign from income 1 column

data['Income.1']=data['Income.1'].str.replace('$','')
data
```

```python
#filling empty spaces with null values

data['Income.1']=data['Income.1'].replace('',np.nan)
data
```

```python

data['Income.1'].isnull().sum()
```

```python
data['Income.1'].dtype
```

```python
#changing data type to float
data['Income.1'] = data['Income.1'].astype(float)
```

```python
data['Income.1'].dtype
```

```python
# using mean of Income.1 column to fill Nan values

data['Income.1']=data['Income.1'].fillna(data['Income.1'].mean())
data
```

```python
# Since there are very high differences between min and max values of Income.2 column
# Hence using median  of Income.2 column to fill Nan values

data['Income.2']=data['Income.2'].fillna(data['Income.2'].median())
data
```

```python
data['Total_income']=data['Income.1']+data['Income.2']
data
```
